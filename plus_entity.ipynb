```python
import pandas as pd
```

```python
# 파일불러오기
ck = pd.read_csv("C:/project/fashion/data/colorkind.csv", header = None, error_bad_lines=False, encoding = 'utf-8')
```

```python
# 컬럼명 변경
ck.rename(columns = {ck.columns[0]:'question'},inplace = True)
ck
```

```python
# '- - '을 없앰
ck['question'] = ck['question'].str.lstrip('- - ')
ck['labels'] = ck['question']
# ck['labels'] = ck['question'].str.split(" ")
```

```python
ck
```

```python
ck['labels'] = ck['labels'].str.replace('색|컬러|네이비|블랙|레드','S-COLOR')
ck['labels'] = ck['labels'].str.replace('종류|원피스|조끼|코트|블라우스|니트|패딩|잠옷|히트텍|복|귀걸이|파카|바지|티|정장|치마|셔츠|남방|잠바|재킷|후드|모자|가방|레깅스|스웨터|스카프|점퍼|양말|맨투맨|타이|운동|바람|머플러|메리야스|슬랙스|스커트|양복|마이|목도리|디건','S-KIND')
```

```python
ck
```

```python
ent=[]
for i in range(len(ck['labels'])):
    ent1 = ck['labels'][i].split(" ")
#     print(ent1)
    ent.append(ent1)
```

```python
ent
```

```python
len(ent)
```

```python
for i in range(len(ent)):
    for j in range(len(ent[i])):
#         if ent[i][j].contains(color):
        if ent[i][j].find('S-COLOR') != -1:
            ent[i][j] = 'S-COLOR'
        elif ent[i][j].find('S-KIND') != -1:
            ent[i][j] = 'S-KIND'
        else:
            ent[i][j] = 'O'
```
