```python
import pandas as pd
import numpy as np
```

```python
ck = pd.read_csv("D:/Workspace/Project_fashion/demo/data/raw/COLORKIND.csv", error_bad_lines=False, encoding = 'utf-8')
cs = pd.read_csv("D:/Workspace/Project_fashion/demo/data/raw/COLORSIZE.csv", error_bad_lines=False, encoding = 'utf-8')
ks = pd.read_csv("D:/Workspace/Project_fashion/demo/data/raw/KINDSIZE.csv", error_bad_lines=False, encoding = 'utf-8')
ps = pd.read_csv("D:/Workspace/Project_fashion/demo/data/raw/PAYSALE.csv", error_bad_lines=False, encoding = 'utf-8')
```

```python
ps
```

```python
# colorkind - color+kind, color
# colorsize - color+size, size
# kindsize - kind+size, kind

ck1 = ck[ck.label.str.contains('S-COLOR|B-COLOR|I-COLOR|E-COLOR')]
cs1 = cs[cs.label.str.contains('S-SIZE|B-SIZE|I-SIZE|E-SIZE')]
ks1 = ks[ks.label.str.contains('S-KIND|B-KIND|I-KIND|E-KIND')]
```

```python
ck1
```

```python
ck1.to_csv('COLORKIND.csv',encoding="euc-kr")
cs1.to_csv('COLORSIZE.csv',encoding="euc-kr")
ks1.to_csv('KINDSIZE.csv',encoding="euc-kr")
```

```python
# intent file
```

```python
ck1.drop(['label'],axis=1, inplace = True)
cs1.drop(['label'],axis=1, inplace = True)
ks1.drop(['label'],axis=1, inplace = True)
ps.drop(['label'],axis=1, inplace = True)
```

```python
ck1['label'] = 'COLORKIND'
cs1['label'] = 'COLORSIZE'
ks1['label'] = 'KINDSIZE'
ps['label'] = 'PAYSALE'
```

```python
ps
```

```python
all_intent = pd.concat([ck1,cs1,ks1,ps],ignore_index=True)
```
